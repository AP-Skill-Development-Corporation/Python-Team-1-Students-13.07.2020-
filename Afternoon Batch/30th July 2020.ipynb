```python
# Merge
import pandas as pd
df1 = pd.DataFrame({"weight":[80,90,60,70],
                   "num":[2,1,2,3],
                   "age":[50,45,45,67]},
                  index=[2001,2002,2003,2004])
df1
```

```python
df2 = pd.DataFrame({"weight":[80,69,60,70],
                   "age":[50,45,45,67],
                   "num":[2,1,2,3]},
                  index=[2001,2004,2005,2006])
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,on='weight')
```

```python
df3 = pd.DataFrame({"weight":[10,29,50,30],
                   "age":[20,15,75,67],
                   "num":[21,1,7,3]},
                  index=[2000,2004,1999,2006])
df3
```

```python
df1.merge(df3)
```

```python
# Join
df4 = pd.DataFrame({"num":[2,1,2,3],
                   "age":[50,45,4,5]},
                  index=[2001,2002,2004,2003])
df4
```

```python
df5 = pd.DataFrame({"num":[2,1,2,3],
                   "age":[50,45,4,5]},
                  index=[2001,2002,2004,2003])
df5
```

```python
df4.join(df5)
```

```python
df6 = pd.DataFrame({"num1":[2,1,2,3],
                   "age1":[50,45,4,5]},
                  index=[2001,2002,2004,2003])
df6
```

```python
df4.join(df6)
```

```python
# concatinate
con = pd.concat([df1,df2])
con
```

```python
# read csv file data
filepath = 'Datafiles/Income.csv'
import pandas as pd
def readCsvdata(filepath):
    return pd.read_csv(filepath)
readCsvdata(filepath)
```

```python
# To print all columns in a single line
incomedata = readCsvdata(filepath)
def PrintDataFrameCols(df):
    columns = df.columns
    for column in columns:
        print(column,end=' ')
    return
PrintDataFrameCols(incomedata)
```

```python
# Average income of all states from 2005-2013
def avgIncone(df):
    for i in range(0,5):
        columns = df.values[i][2:]
        total = 0
        count = 0
        for column in columns:
            total = total + column
            count = count + 1
        print(df.values[i][1],total//count)
    return
incomedata = readCsvdata(filepath)
avgIncone(incomedata)
```

```python
# Basic Linear Graph
import matplotlib.pyplot as plt
plt.title('Basic linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Non linear graph
# v = at
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i]*i for i in range(len(t))]
plt.plot(t,v,'gx')
plt.title("Non Linear Graph")
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.grid()
plt.show()
```

```python
# Bar graph
x = [1,2,3,4]
y = [10,20,30,40]
plt.bar(x,y)
plt.show()
```

```python
# pie 
x=[1,2,3,4]
y=[10,20,30,40]
plt.pie(x)
plt.title('basic pie ')
plt.show()
```
