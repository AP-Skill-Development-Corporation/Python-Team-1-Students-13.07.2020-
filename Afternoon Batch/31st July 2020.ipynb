```python
import matplotlib.pyplot as plt
xaxis = [25,26,27,28]
yaxis=[2800,2900,3000,2100]
plt.plot(xaxis,yaxis,color='k',linestyle='--',marker='o',linewidth=2)
y1axis = [2000,4000,9000,1000]
plt.plot(xaxis,y1axis,color = '#5a7d9a',marker='d',label='y1axis')
plt.title('Basic plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.style.use('fivethirtyeight')
plt.grid(True)
#print(plt.style.available)
plt.show()
```

```python
import numpy as np
# plot 1d numpy arrays
x = np.linspace(10,100)
y = np.linspace(5,100)
print(x)
print(y)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
# lim,ylim means x,y limits
plt.xlim([20,80])
plt.ylim([200,800])
plt.grid(False)
plt.show()
```

```python
# time vs distance
distance = list(range(2,11,2))
time = list(range(1,6))
plt.plot(time,distance,'bo')
plt.show()
```

```python
dist= np.linspace(5,100)
speed= np.linspace(20,100)
k = dist/speed
j = plt.plot(k)
plt.title('Distance Vs Speed')
plt.xlabel('distance')
plt.ylabel('Speed')
plt.show()
plt.savefig('speedgraph.png')
```

```python
# sinusoidal graph
import numpy as np
t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.title('Sine wave')
plt.plot(t,v,'go')
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Sine-wave.png')
plt.show()
```

```python
# figures and subplots
x = np.linspace(1,10,100)
y = np.log(x)
plt.figure(1)
plt.subplot()
plt.title('y=log(x)')
plt.plot(x,y)

plt.figure(2)
plt.subplot()
plt.title('y=log(x)**2')
plt.plot(x,y**2)
plt.show()
```

```python
x = np.linspace(1,10,100)
plt.figure(1)
plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)

plt.show()
```

```python
xaxis = [25,26,27,28]
yaxis=[2800,2900,3000,2100]
plt.plot(xaxis,yaxis,color='k',linestyle='--',marker='o',linewidth=2)
y1axis = [2000,4000,9000,1000]
plt.plot(xaxis,y1axis,color = '#5a7d9a',marker='d',label='y1axis')
plt.fill_between(xaxis,yaxis,y1axis,color='red',alpha=0.25)
plt.title('Fill between plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.style.use('fivethirtyeight')
plt.grid(True)
#print(plt.style.available)
plt.show()
```

```python
# plot using csv file
import pandas as pd
df = pd.read_csv('Datafiles/Income.csv')
years = df.columns[2:]
incomeArkanas = df.values[3,2:]
plt.title("Arkanas Income")
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkanas)
plt.grid(True)
plt.show()
```

```python
# Income of all states
#df
incomeallstates = []
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title("Income of all states ")
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```
