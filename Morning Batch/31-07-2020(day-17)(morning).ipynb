# matplotlib

```python
- raw data converts into graphical representation.
- datavisualization
- helps us to understand data clearly and effectively.
- pip install matplotlib
```

```python
# format strings
- color,linestyle,marker
```

```python
#basic linear plot
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y=[28000,29000,30000,31000,32000,33000]
plt.plot(x,y,label='y_axis',color='y',linestyle='--',marker='d')
y1=[28000,32000,49000,50000,65000,75000]
plt.plot(x,y1,label='y1_axis',color='#5a7d9a',linestyle='-.',marker='^')
plt.tight_layout()
plt.title('basic linear plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.grid(True)
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

```python
import numpy as np
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)
plt.plot(x,y,'yo',x,y**2,'r--',x,y**3,'b+',x,y**4,'g^')
plt.show()
```

```python
x=np.linspace(1,10,100)
plt.figure(1)
#subplot1
plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

#subplot2
plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)
#subplot3
plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

#subplot4
plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)

plt.show()
```

```python
# bar graph
x=[25,26,27,28,29,30]
y=[60000,29000,30000,40000,41000,50000]
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```

```python
# barh graph
x=[25,26,27,28,29,30]
y=[60000,29000,30000,40000,41000,50000]
plt.barh(x,y)
plt.title('barh graph')
plt.show()
```

```python
# pie chart
labels=['python','html','java','c','sql']
slices=[60000,29000,30000,40000,41000]
explode=[0,0,0.1,0,0]
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},startangle=90,shadow=True)
plt.title('pie chart')
plt.style.use('fivethirtyeight')
plt.show()
```
